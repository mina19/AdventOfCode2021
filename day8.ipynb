```python
mydict = {1: set(['C','F']),
          7: set(['A','C','F']),
          4: set(['B','C','D','F']),
          2: set(['A','C','D','E','G']),
          3: set(['A','C','D','F','G']),
          5: set(['A','B','D','F','G']),
          0: set(['A','B','C','E','F','G']),
          6: set(['A','B','D','E','F','G']),
          9: set(['A','B','D','D','F','G']),
          8: set(['A','B','C','D','E','F','G'])}
mydict2 = dict(
    map(lambda item: (item[0], len(item[1])), mydict.items())
)
mydict2r = {2: 1, 4:4, 3:7, 7:8}
```

```python
count = 0
decode = {}
nonunique = {i:[] for i in [0,2,3,5,6]}
final = []
for i in range(len(input)):
    line = input[i]
    decode = {}
    nonunique = {i:[] for i in [0,2,3,5,6]}
    # six in nonunique[6] has to be 0,6,9 in some order
    # five in nonunique[5] has to be 2,3,5 in some order
    line
    for signal in line.split(' '):
        length = len(signal)
        if length in mydict2r.keys():
            count += 1
            #print('{}: {}'.format(signal, mydict2r[length]))
            decode[mydict2r[length]] = set(signal)
        else:
            nonunique[length].append(set(signal))
    decode.keys()
    for six in nonunique[6].copy():
        if list(decode[8] - six)[0] in decode[1]:
            C = list(decode[8] - six)[0]
            F = list(decode[1] - set(C))[0]
            decode[6] = six
            nonunique[6].remove(six)
    for five in nonunique[5].copy():
        if (len(decode[6] - five))==1:
            E = list(decode[6] - five)[0]
            decode[5] = five
            nonunique[5].remove(five)
            decode[9] = decode[5] | decode[1]
            nonunique[6].remove(decode[9])
            decode[0] = nonunique[6][0]
            nonunique[6].remove(decode[0])
    for five in nonunique[5].copy(): #only two and three in nonunique[5] remain. decode[7] is in decode[3]
        if len(set(five) - set(decode[7]))==2:
            decode[3] = five
            nonunique[5].remove(five)
        else:
            decode[2] = five
            nonunique[5].remove(five)
    encode = dict(map(lambda item: (item[0], list(item[1])), decode.items()))
    encode = dict(map(lambda item: (item[0], sorted(item[1])), encode.items()))
    encode = dict(map(lambda item: (item[0], ''.join(list(item[1]))), encode.items()))
    encode = dict(map(lambda item: (item[1], item[0]), encode.items()))
    
    line = output[i]
    nums = ''
    for signal in line.split(' '):
        num = sorted(list(signal))
        num = ''.join(num)
        nums += str(encode[num])
    print(nums)
    final.append(int(nums))
```

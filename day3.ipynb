```python
import pandas as pd
import numpy as np
```

```python
with open("day3ex2.txt", 'r') as f:
    data = f.read().splitlines()
```

```python
mydict = {}
for i in range(len(data)):
    mydict[i] = [int(char) for char in data[i]]
data = pd.DataFrame.from_dict(mydict)
```

```python
data = pd.DataFrame.from_dict(mydict)
dataT = data.T
```

```python
data
```

```python
gamma_rate = dataT.mode().iloc[0].ravel()
```

```python
gamma_rate
```

```python
epsilon_rate = np.where(gamma_rate==1, -1, gamma_rate)
epsilon_rate = np.where(epsilon_rate==0, 1, epsilon_rate)
epsilon_rate = np.where(epsilon_rate==-1, 0, epsilon_rate)
epsilon_rate, gamma_rate
```

```python
gamma, epsilon = 0, 0
def convert(mylist):
    result = 0
    for digit in mylist:
        result = (result << 1) | digit
    return result
gamma = convert(gamma_rate)
epsilon = convert(epsilon_rate)
gamma*epsilon
```

```python
# part2
dataT2 = dataT.copy()
for i in range(dataT2.shape[1]):
    mode=dataT2.mode()[i].ravel()
    if len(mode)==2:
        mode = 1
    else:
        mode = mode[0]
    dataT2 = dataT2[dataT2[i]==mode]
    if dataT2.shape[0] == 1:
        break
oxygen_gen = dataT2.values[0]
oxygen = convert(oxygen_gen)
oxygen
```

```python
# part2
dataT2 = dataT.copy()
for i in range(dataT2.shape[1]):
    mode=dataT2.mode()[i].ravel()
    if len(mode)==1:
        mode = 0 if mode==1 else 1
    else:
        mode = 0
    dataT2 = dataT2[dataT2[i]==mode]
    if dataT2.shape[0] == 1:
        break
co2_gen = dataT2.values[0]
co2 = convert(co2_gen)
co2
```

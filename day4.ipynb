```python
with open('day4ex2.txt', 'r') as f:
    data = f.read().splitlines()
    
draw_num = [int(char) for char in data[0].split(',')]
data = [i for i in data if i != '']

boards = {}
board_num = (len(data)-1) / 5
data2=[]

for i in range(1,len(data)):
    data2.append(data[i].split(' '))
for i in range(len(data2)):
    info = [int(j) for j in data2[i] if j!= '']
    board = i//5
    if board in boards:
        boards[board].append(info)
    else:
        boards[board] = [info]
```

```python
def check_board(board):
    # go across rows
    for j in range(5):
        res=[True if i==True else False for i in board[j]]
        res = res[0] and res[1] and res[2] and res[3] and res[4]
        if res == True:
            return True
    # go across columns
    for j in range(5):
        row = [item[j] for item in board]
        res=[True if i==True else False for i in row]
        res = res[0] and res[1] and res[2] and res[3] and res[4]
        if res == True:
            return True
    return False

def tot(board):
    tot = 0
    for i in range(5):
        row = [0 if i==True else i for i in board[i]]
        tot += sum(row)
    return tot
```

```python
def compute(boards):
    for j in draw_num:
        for board in boards:
            for m in range(5):
                boards[board][m] = [True if char==j else char for char in boards[board][m]]
            if check_board(boards[board])==True:
                return(j*tot(boards[board]))

compute(boards)
```

```python
def compute_lose(boards):
    boardstrack = {i:[] for i in range(len(boards))}
    for j in draw_num:
        for board in boards:
            for m in range(5):
                boards[board][m] = [True if char==j else char for char in boards[board][m]]
            if check_board(boards[board])==True:
                if boardstrack[board]==[]:
                    print(board)
                    boardstrack[board] = j*tot(boards[board])
                else:
                    pass
    return boardstrack
                
compute_lose(boards)
```
